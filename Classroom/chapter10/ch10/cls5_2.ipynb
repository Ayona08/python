```python
class Card:
    FACES=['Ace','2','3','4','5','6','7','8','9','10','King','Queen','Jack']
    SUITS=['Hearts','Spades','Diamonds','Clubs']
    def __init__(self,face,suit):
        self.face=face
        self.suit=suit
    def __str__(self):
        return f'{self.face} of {self.suit}'
    def image_name(self):
        return str(self).replace(' ','_')+'.png'
c1=Card('Jack','Clubs')
print(c1.face)
print(c1.suit)
print(Card.FACES)
print(Card.SUITS)
print(c1.image_name())        
```

```python
import random as r
class DeckOfCard:
    NUMBER_OF_CARDS=52
    def __init__(self):
        self.current_card=0
        self.card_list=[]
        for index in range(DeckOfCard.NUMBER_OF_CARDS):
            face=Card.FACES[index%13]
            suit=Card.SUITS[index//13]
            self.card_list.append(Card(face,suit))
    def deal_card(self):
        try:
            card=self.card_list[self.current_card]
            self.current_card+=1
            return card
        except IndexError:
            return None
    def shuffle(self):
        r.shuffle(self.card_list)
    def __str__(self):
        s=''
        for i in range(52):
            s+=f'{str(self.card_list[i]):<20}'
            if(i+1)%4==0:
                s+='\n'
        return s
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from pathlib import Path
d = DeckOfCard()
d.shuffle()
path = Path('.').joinpath('card_images') 
fig, axes_list = plt.subplots(nrows=13, ncols=4, figsize=(10, 12))
for ax, card in zip(axes_list.ravel(), d.card_list):
    ax.set_axis_off()
    image_name = card.image_name() 
    img_path = path.joinpath(image_name)
    img = mpimg.imread(img_path) 
    ax.imshow(img)
plt.tight_layout()
plt.show()
```

```python
#ce.py
class CommisionEmployee(object):
    def __init__(self,f_name,l_name,ssn,gross_sales,commision_rate):
        self.f_name=f_name
        self.l_name=l_name
        self.ssn=ssn
        self.gross_sales=gross_sales
        self.commision_rate=commision_rate
    @property
    def gross_sales(self):
        return self._gross_sales
    @gross_sales.setter
    def gross_sales(self,value):
        if value<0:
            raise ValueError("GS can not be negative")
        self._gross_sales=value
    @property
    def commision_rate(self):
        return self._commision_rate
    @commision_rate.setter
    def commision_rate(self,value):
        if not(0<value<1):
            raise ValueError("value must be between 0 - 1")
        self._commision_rate=value
    def earning(self):
        return self.gross_sales*self.commision_rate
    def __str__(self):
        return f'CommisionEmployee:\n'+\
               f'Name:{self.f_name} {self.l_name}\n'+\
               f'Earning: {self.earning()}'

```

```python
class SalariedCommisionEmployee(CommisionEmployee):
    def __init__(self,f_name,l_name,ssn,gross_sales,commison_rate,base_salary):
        super().__init__(f_name,l_name,ssn,gross_sales,commison_rate)
        self.base_salary=base_salary
    def earning(self):
        return super().earning()+self.base_salary
    def __str__(self):
        return f'Salaried{super().__str__()}'



class Duck:
    def __init__(self__):
        pass
    def earning(self):
        return f'earning is 100000'
    def __str__(self):
        return f'i am well paid Duck'



    
```
